## Regressions

* Perform a series of regressions on various combinations of the independent variables.
* Plot the predicted values versus the true values
* Which combinations gives the best fit?

You can use either scikit-learn or statsmodels

```python
# Plotting the predicted models versus the true values
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm

cols2 = ['beds', 'baths', 'sq__ft', 'latitude', 'longitude']

lm = linear_model.LinearRegression()
def func1_w_plots(i):
    X = data[i]
    y = data['price']

    model = lm.fit(X, y)
    predictions = lm.predict(X)

    # Plot the model
    plt.scatter(predictions, y, s=30, c='r', marker='+', zorder=10)
    plt.xlabel("Predicted Values Price")
    plt.ylabel("Actual Values Price")
    plt.show()
    print i
    print mean_squared_error(y, predictions)
    return


print [func1_w_plots([i]) for i in cols2]

```

```python
# Finding the best combinations of variables
lm = linear_model.LinearRegression()
def func1(i):
    X = data[i]
    y = data['price']
    model = lm.fit(X, y)
    predictions = lm.predict(X)
    return mean_squared_error(y, predictions)
    
def find_best(i):
    best = {}
    for x in i:
        best[x] = func1([x])
    print (min(best, key=best.get)), best[(min(best, key=best.get))]
print 'Best fit with one variable:',find_best(cols2)
    
def find_best_2(i):
    two_best = {}
    for x in i:
        for j in i:
            if x != j:
                two_best[x, j] = func1([x, j])
    print (min(two_best, key=two_best.get)), two_best[(min(two_best, key=two_best.get))]
    return
print 'Best fit with two variables:',find_best_2(cols2)

def find_best_3(i):
    three_best = {}
    for x in i:
        for j in i:
            for k in i:    
                if x != j != k:
                    three_best[x, j, k] = func1([x, j, k])
    print (min(three_best, key=three_best.get)), three_best[(min(three_best, key=three_best.get))]
    return
print 'Best fit with three variables:',find_best_3(cols2)

def find_best_4(i):
    four_best = {}
    for x in i:
        for j in i:
            for k in i:    
                for t in i:    
                    if x != j != k != t:
                        four_best[x, j, k, t] = func1([x, j, k, t])
    print (min(four_best, key=four_best.get)), four_best[(min(four_best, key=four_best.get))]
    return
    
print 'Best fit with four variables:', find_best_4(cols2)
```
